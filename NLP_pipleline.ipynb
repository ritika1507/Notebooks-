# STEP 4  Lemmatization

It is the process of reducing inflected forms of a word while still ensuring that the reduced form belongs to the language.
This reduced form or root word is called a lemma.

```python
#printing those lemma which are different from the actual word


print([{token , token.lemma_} for token in doc if str(token) != (token.lemma_)])
```

#STEP 6 : Dependency parsing
The goal is to build a tree that assigns a single parent word to each word in the sentence. The root of the tree will be the main verb in the sentence.

## Additional Exercises for practice

#Replace a token with '"REDACTED" if its a name

```python
#removing names from the document



def replace_name(token):
  if token.ent_iob != 0 and token.ent_type_ == "PERSON":
    return "[REDACTED]"
  else:
     return token.text


#function to remove names and print doc
def modify_doc(doc):
  with doc.retokenize() as retokenizer:
    for ent in doc.ents:
      retokenizer.merge(ent)
  tokens = map(replace_name , doc)
  return " ".join(tokens)


print(modify_doc(doc))

#names have been replaced with the word "REDACTED"
```

## Extracting facts from text

```python
import textacy.extract
```

```python
#extract semistructured statements

statements = textacy.extract.semistructured_statements(doc , entity = "Chess" , cue = "be")

print("Lets get to know something about chess:")

# print(statements)
for line in statements:
  subject , verb , fact = line
  print(f"- {fact}")
```

#Getting to know token attributes

```python

for token in doc:
  print([token , token.idx , token.text_with_ws , token.is_alpha , token.is_punct , token.is_space, 
        token.shape_ , token.is_stop])
  

#text_with_ws prints token text with trailing space (if present).
# is_alpha detects if the token consists of alphabetic characters or not.
# is_punct detects if the token is a punctuation symbol or not.
# is_space detects if the token is a space or not.
# shape_ prints out the shape of the word.
# is_stop detects if the token is a stop word or not.
```

# Word frequency & common words

```python


from collections import Counter
words = [token.text for token in doc if not token.is_stop and not token.is_punct]
word_freq = Counter(words)
common_words = word_freq.most_common(10)
print (common_words)
```

## Removing stop words and punctuation tokens

```python
def is_token_allowed(token):
  if (not token or token.is_stop or token.is_punct):
    return False
  return True


def preprocess_token(token):
  return token.lemma_.strip().lower()
     
     
complete_filtered_tokens = [preprocess_token(token)
  for token in doc if is_token_allowed(token)]
print(complete_filtered_tokens)
```
