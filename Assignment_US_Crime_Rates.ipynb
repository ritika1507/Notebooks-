### Step 3. Assign it to a variable called crime.

```python
crime = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/US_Crime_Rates/US_Crime_Rates_1960_2014.csv')
```

```python
crime.head()
```

### Step 4. What is the type of the columns?

```python
crime.dtypes
```

##### Have you noticed that the type of Year is int64. But pandas has a different type to work with Time Series. Let's see it now.

### Step 5. Convert the type of the column Year to datetime64

```python
crime['Year'] = pd.to_datetime(crime['Year'] , format = ('%Y'))
crime.head()
```

```python
crime.dtypes
```

```python
crime.shape
```

### Step 6. Set the Year column as the index of the dataframe

```python
crime_idx = crime.set_index('Year')
crime_idx.head()
```

### Step 7. Delete the Total column

```python
crime_idx = crime_idx.drop('Total' , axis = 1)
```

```python
crime_idx.head()
```

### Step 8. Group the year by decades and sum the values

#### Pay attention to the Population column number, summing this column is a mistake

```python

#including all columns and summing them up except for the population column
crime_decades = crime_idx.groupby((crime_idx.index.year//10)*10)[crime_idx.iloc[: , 1 : ].columns].sum()
crime_decades.head()
```

### Step 9. What is the most dangerous decade to live in the US?

```python
crime_decades.idxmax(axis = 0)
```

We can observe from here that the most dangerous decade to live in the US was 1990s.
