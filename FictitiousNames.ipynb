### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Create the 3 DataFrames based on the following raw data

```python
raw_data_1 = {
        'subject_id': ['1', '2', '3', '4', '5'],
        'first_name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'], 
        'last_name': ['Anderson', 'Ackerman', 'Ali', 'Aoni', 'Atiches']}

raw_data_2 = {
        'subject_id': ['4', '5', '6', '7', '8'],
        'first_name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'], 
        'last_name': ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']}

raw_data_3 = {
        'subject_id': ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],
        'test_id': [51, 15, 15, 61, 16, 14, 15, 1, 61, 16]}
```

### Step 3. Assign each to a variable called data1, data2, data3

```python
data1 = pd.DataFrame.from_dict(raw_data_1)
data1.head()
```

```python
data2 = pd.DataFrame.from_dict(raw_data_2)

data2.head()
```

```python
data3 = pd.DataFrame.from_dict(raw_data_3)
data3.head()
```

### Step 4. Join the two dataframes along rows and assign all_data

```python
#concatinating the datasets
all_data = pd.concat([data1 , data2]) 
all_data
```

```python

#ignoring index as they are not significant and 
#have values now from 0-9 
all_data = pd.concat([data1 , data2] , ignore_index = True)
all_data
```

### Step 5. Join the two dataframes along columns and assing to all_data_col

```python
#joining along columns

all_data_col = pd.concat([data1 , data2] , axis = 1 )
all_data_col
```

### Step 6. Print data3

```python
data3
```

### Step 7. Merge all_data and data3 along the subject_id value

```python
#merging all_data and data3
#on subject_id

new_data = pd.merge(all_data , data3 , on = ['subject_id'])
new_data
```

### Step 8. Merge only the data that has the same 'subject_id' on both data1 and data2

```python
#inner join on data1 , data2

new_data2 = pd.merge(data1 , data2 , on = 'subject_id')
new_data2
```

### Step 9. Merge all values in data1 and data2, with matching records from both sides where available.

```python
#outer join to get matching records from both tables on the common column subject_id
new_data3 = pd.merge(data1 , data2 , on = 'subject_id' , how = 'outer')
new_data3
```
